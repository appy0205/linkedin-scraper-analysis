```python
grouped_single = df.groupby('Gender').agg({'Age': ['mean', 'min', 'max']})

# Extract the mean ages for plotting
mean_age = grouped_single['Age']['mean']

# Create a bar plot
mean_age.plot(kind='bar')
plt.title('Mean Age by Gender')
plt.xlabel('Gender')
plt.ylabel('Mean Age')
plt.xticks(rotation=0)  # Rotate x-axis labels if needed
plt.show()
```
